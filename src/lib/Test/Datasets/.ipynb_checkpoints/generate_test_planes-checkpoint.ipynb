```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.spatial.distance import cdist
from mpl_toolkits.mplot3d import Axes3D
import pandas as pd
```

```python
L = 100
W = 50
kx = 11
ky = 11
```

```python
x, y = np.mgrid[0:L:kx*1j, 0:W:ky*1j]
```

```python
n = np.array([1, 0, 1])
n = n / np.linalg.norm(n)
D = 0
n, D
```

```python
A, B, C = n
```

```python
z = -(A*x+B*y+D)/C
z
```

```python
fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x, y, z, alpha=0.5)
ax.scatter(x, y, z, alpha=0.5)
```

```python
ix, iy = np.mgrid[0:kx, 0:ky]
```

```python
df = pd.DataFrame({'tr':ix.ravel(), 't':iy.ravel(), 'x':x.ravel(), 'y':y.ravel(), 'z':z.ravel()})
#df.to_csv('test_plane_1.csv', index=False)
```

```python
def make_dotted_plane(n, D=0):
    n = n / np.linalg.norm(n)
    x, y = np.mgrid[0:L:kx*1j, 0:W:ky*1j]
    A, B, C = n
    z = -(A*x+B*y+D)/C
    ix, iy = np.mgrid[0:kx, 0:ky]
    return pd.DataFrame({'tr':ix.ravel(), 't':iy.ravel(), 'x':x.ravel(), 'y':y.ravel(), 'z':z.ravel()})

def show_plane(*planes):
    fig = plt.figure(figsize=(10, 10))
    ax = fig.add_subplot(111, projection='3d')
    for plane in planes:
        x, y, z = plane['x'], plane['y'], plane['z']
        ax.scatter(x, y, z, alpha=0.5)
        #ax.plot_surface(x, y, z, alpha=0.5)
```

```python
df1 = make_dotted_plane(np.array([1, 0, 1]))
df2 = make_dotted_plane(np.array([1, 1, 1]))
show_plane(df1, df2)
```
